  2. A) Lasso Regression is a type of linear regression that incorporates a regularization penalty into the estimation process. Lasso helps prevent overfitting, which is a common problem in such scenarios, by adding a penalty that encourages simpler, more interpretable models that generalize better on new data. The formula for Lasso Regression can be expressed as:
The formula for Lasso Regression can be expressed as:
$\min_{\beta} \left\{ \frac{1}{2n} \sum_{i=1}^n (y_i - X_i \beta)^2 + \lambda \sum_{j=1}^p |\beta_j| \right\}$
Here's the breakdown of the components in the Lasso Regression formula:

- $\beta$ represents the vector of coefficients, including the intercept.
- $y_i$ are the observed values.
- $X_i$ are the predictor variables.
- $n$ is the number of observations.
- $p$ is the number of predictors.
- $\lambda$ is a non-negative regularization parameter that controls the strength of the penalty. As $\lambda$ increases, the penalty for non-zero coefficients becomes more severe, driving more coefficients to zero.

  A) Double Lasso Regression helps clarify the true effect of one variable on another in two main steps, managing too many data points effectively:

### Step 1: Choosing Important Factors

**What it does:** This step selects the important factors (like background details) that might influence the main variable we're studying.

**Formula:**
\[ \min_{\gamma} \left\{ \frac{1}{2n} \sum_{i=1}^n (X_i - W_i \gamma)^2 + \lambda_1 \sum_{j=1}^p |\gamma_j| \right\} \]

- **\(X_i\)**: The main variable of interest.
- **\(W_i\)**: All possible factors that could influence \(X_i\).
- **Purpose**: To identify which factors significantly predict \(X_i\) by using Lasso, which simplifies the model by eliminating less important factors.

### Step 2: Estimating the Real Impact

**What it does:** Using the important factors identified from Step 1, this step calculates the actual impact our main variable has on the outcome.

**Formula:**
\[ \min_{\beta} \left\{ \frac{1}{2n} \sum_{i=1}^n (Y_i - X_i \beta - W_i \delta)^2 + \lambda_2 \sum_{k=1}^{p'} |\delta_k| \right\} \]

- **\(Y_i\)**: The outcome we want to understand.
- **\(X_i\)**: The main variable from Step 1.
- **Purpose**: To accurately measure how much \(X_i\) affects \(Y_i\), adjusting for those important factors we found earlier to ensure we capture the true causal effect.

  1. Traditional RNNs struggle with learning from long data sequences because their memory is short-lived, often forgetting earlier input as new data comes in. LSTMs, a special kind of RNN, solve this by using a clever system of gates that manage information: they decide what to keep, what to discard, and what to pass on. This helps LSTMs remember important information for much longer, making them much better at handling tasks like understanding sentences or predicting future events in a series. Essentially, LSTMs can learn and remember more effectively over long sequences than traditional RNNs.
  2. An LSTM unit is a type of neural network designed to handle long sequences of data without forgetting important information. It has a central component called the cell state that carries data down the sequence chain, acting like a memory line. The LSTM also uses three gates to manage this information: the forget gate decides which data to drop from the cell state, the input gate adds new data to it, and the output gate controls what gets sent out of the LSTM at each step. These gates help the LSTM keep or discard information, making sure it remembers important data even in long sequences. This design helps LSTMs perform well in tasks where understanding the connection between distant events in data is crucial.
  3.
In an LSTM, the **sigmoid (σ)** and **hyperbolic tangent (tanh)** activation functions are crucial for controlling information flow. The sigmoid function is defined as \( \sigma(x) = \frac{1}{1 + e^{-x}} \) and outputs values between 0 and 1. This function helps the gates decide how much information to let through, for example, helping the forget gate determine how much of the past information to keep or discard.

The **tanh** function, expressed as \( \tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}} \), outputs values between -1 and 1 and is used to scale the information being added to the cell state. This ensures that the values in the cell state stay within a manageable range, normalizing the effect of any updates.

Together, these functions ensure that the LSTM can effectively update and maintain its cell state, enabling it to retain or forget information as needed for tasks that require understanding long sequences.

```python
import pandas as pd

# Sample data
data = {
    'Country': ['USA', 'USA', 'USA', 'Canada', 'Canada', 'Canada', 'UK', 'UK', 'UK'],
    'Year': [2019, 2020, 2021, 2019, 2020, 2021, 2019, 2020, 2021],
    'GDP': [1000, 1200, pd.NA, 800, pd.NA, 1000, pd.NA, 900, 1100],
    'InflationRate': [2.0, pd.NA, 3.5, 1.8, 2.1, pd.NA, 2.2, pd.NA, 2.5],
    'UnemploymentRate': [5.0, 4.8, 5.2, pd.NA, 6.0, 5.5, 4.0, 3.8, pd.NA]
}

# Create DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
print("Original DataFrame:")
print(df)

# Group the data by country and year and fill missing values with the column means within each group
df_filled = df.groupby(['Country', 'Year']).apply(lambda x: x.fillna(x.mean()))

# Display the DataFrame after filling missing values
print("\nDataFrame after filling missing values:")
print(df_filled)

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    'Year': range(2012, 2023),
    'Dominican Republic GDP': [1000, 1100, 1200, 1300, 1350, 1400, 1450, 1500, 1550, 1600, 1650],
    'Puerto Rico GDP': [800, 820, 840, 870, 890, 910, 930, 950, 970, 990, 1010]
}

# Create DataFrame
df = pd.DataFrame(data)

# Set 'Year' as index
df.set_index('Year', inplace=True)

# Plot GDP growth for Dominican Republic and Puerto Rico
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['Dominican Republic GDP'], marker='o', label='Dominican Republic')
plt.plot(df.index, df['Puerto Rico GDP'], marker='o', label='Puerto Rico')
plt.title('GDP Growth Over the Last Decade')
plt.xlabel('Year')
plt.ylabel('GDP')
plt.xticks(df.index)
plt.grid(True)
plt.legend()
plt.show()

# Calculate mean, median, and standard deviation of GDP growth rates
mean_dominican_rep = df['Dominican Republic GDP'].pct_change().mean()
median_dominican_rep = df['Dominican Republic GDP'].pct_change().median()
std_dominican_rep = df['Dominican Republic GDP'].pct_change().std()

mean_puerto_rico = df['Puerto Rico GDP'].pct_change().mean()
median_puerto_rico = df['Puerto Rico GDP'].pct_change().median()
std_puerto_rico = df['Puerto Rico GDP'].pct_change().std()

print("\nGDP Growth Statistics for Dominican Republic:")
print("Mean: ", mean_dominican_rep)
print("Median: ", median_dominican_rep)
print("Standard Deviation: ", std_dominican_rep)

print("\nGDP Growth Statistics for Puerto Rico:")
print("Mean: ", mean_puerto_rico)
print("Median: ", median_puerto_rico)
print("Standard Deviation: ", std_puerto_rico)
```
